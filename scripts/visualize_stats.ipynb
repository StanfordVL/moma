# Dataset Statistics Visualization

```python
from IPython.display import Image, display
from momaapi import MOMA, StatVisualizer
```

Initialize MOMA API.

```python
dir_moma = "/home/alan/data/moma-lrg"
moma = MOMA(dir_moma)
```

```python
visualizer = StatVisualizer(moma)
names = {"act": "Activity", "sact": "Sub-activity",
         "actor": "Actor", "object": "Object",
         "att": "Attribute", "rel": "Relationship"}
```

## Overall Statistics

```python
%%capture
paths = visualizer.show(with_split=False)
```

```python
for key, path in paths.items():
    print(names[key])
    display(Image(path))
```

## Statistics by Split

```python
%%capture
paths = visualizer.show(with_split=True)
```

```python
for key, path in paths.items():
    print(names[key])
    display(Image(path))
```
